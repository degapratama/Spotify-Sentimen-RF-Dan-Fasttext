```python
from google_play_scraper import Sort, reviews
import pandas as pd

result, _ = reviews(
    'com.spotify.music', 
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=4000
)

df = pd.DataFrame(result)
df = df[['userName', 'score', 'at', 'content']]

df.to_csv('Dataset/raw_review_2.csv', index=False)
df.head()
```
